```python
import numpy as np
import matplotlib.pyplot as plt

#Generate data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

#Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

#Calculate thew predicted values of y using the regression line
y_pred = slope * x + intercept

#Plot the original data andpoints and the regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Generate data
x = np.array([60, 90, 70, 85, 80, 65, 75, 60, 75, 70])
y = np.array([70, 95, 70, 75, 90, 75, 75, 60, 80, 65])

#Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

#Calculate thew predicted values of y using the regression line
y_pred = slope * x + intercept

#Create title and label for X and Y
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")

#Plot the original data andpoints and the regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```

The sample data problem having a r = 0.7879 which has a strong postive linear correlation, indicates that the line is upward that is shown visually as we can see above that is given from the plotted data points and the positive line regression.
