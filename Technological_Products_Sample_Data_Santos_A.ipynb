```python
import matplotlib.pyplot as plt

#3. Analyze the data and generate the appropriate plot includingits properties:
#a. Read each device's date of release and sales. Show it using line plot.

plt.figure(figsize=(18, 7)) #decalred as to show a wider view for the plot. (User's preference number)

x_label = techno['Date of Release']
y_label = techno['Price (USD)']
plt.plot(x_label,y_label)
plt.title("Date of release and sales")


plt.xlabel("Date of Release")
plt.ylabel("Price (USD)")
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'Date of Release' to datetime and sort the DataFrame
techno['Date of Release'] = pd.to_datetime(techno['Date of Release'], errors='raise')
techno = techno.sort_values('Date of Release')

# Create a figure for the plot
plt.figure(figsize=(18, 7))

# Plotting with a single marker 'o'
for brand in techno['Brand'].unique():
    brand_data = techno[techno['Brand'] == brand]
    plt.plot(brand_data['Date of Release'], brand_data['Price (USD)'],
             marker='o', label=brand)

# Adding titles and labels
plt.title('Device Prices Over Release Dates', fontsize=16)
plt.xlabel('Date of Release', fontsize=16)
plt.ylabel('Price (USD)', fontsize=16)
plt.xticks(rotation=50, ha='right')
plt.legend(fontsize=10)
plt.grid(True)

plt.grid()
plt.show()
```

```python
#.d Read each brand sales data and show it using the bar chart. Label them.

import matplotlib.pyplot as plt

plt.figure(figsize=(18, 7))
x = techno['Brand']
y = techno['Price (USD)']

plt.title("Total Sales by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Sales (USD)")
plt.bar(x, y)

plt.xticks(rotation=50, ha='right')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# e. Read each device and show it using the histogram to see the most common Country of Origin. Label them.

plt.figure(figsize=(18, 7))
data = techno['Country of Origin']
plt.hist(data, bins=data.nunique(), edgecolor ='black')
plt.title("Histogram of Country of Origin for Devices")
plt.xlabel("Country of Origin")
plt.ylabel("Devices")

plt.xticks(rotation = 50)
plt.grid(axis = 'y', linestyle = '--')
plt.show()
```

```python
import matplotlib.pyplot as plt

brand_countries = techno[['Brand', 'Country of Origin']].drop_duplicates()

plt.figure(figsize=(8, 8))
plt.pie(brand_countries['Country of Origin'].value_counts(),
        labels=brand_countries['Country of Origin'].value_counts().index,
        autopct='%1.1f%%', startangle=90,
        textprops={'fontsize': 18})

plt.title('Brand Country of Origin')
plt.show()
```
