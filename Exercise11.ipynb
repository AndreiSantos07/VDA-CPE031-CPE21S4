```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/drive/MyDrive/Sample Data (Exercise 11) - Sheet1.csv'
df = pd.read_csv('/content/drive/MyDrive/Sample Data (Exercise 11) - Sheet1.csv')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Generate data
x = np.array([0.50, 1.00, 1.30, 1.60, 2.00, 2.20, 2.50, 3.10, 3.90, 4.40])
y = np.array([1.90, 1.50, 1.20, 1.00, 0.80, 0.78, 0.65, 0.46, 0.30, 0.23])

slope, intercept = np.polyfit(x, y, 1)
y_pred = slope * x + intercept

plt.title("Linear Model of the Sample Problem")
plt.xlabel("X")
plt.ylabel("Y")

plt.scatter(x, y)
plt.plot(x, y_pred, color = 'green', marker = linewidth = 3, markerfacecolor='red')
plt.show()
```

The sample data problem having a r = -0.9475 which has a strong negative linear correlation, indicating that this relationship is visually shown in the downward part of the plotted data points and the negatively sloped regression line.
