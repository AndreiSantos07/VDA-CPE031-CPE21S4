Question:
How does the average number of mental health days differs the various general health statuses?

```python
import matplotlib.pyplot as plt

genhp_over_mtlhp = heart1.groupby('GeneralHealth')['MentalHealthDays'].mean()

plt.figure(figsize=(18, 7))
genhp_over_mtlhp.plot.bar(color = 'teal')
plt.title('Average Mental Health Days by General Health Status')
plt.xlabel('General Health Status')
plt.ylabel('Average Mental Health Days')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha = 0.56)
plt.show()
```

Observation:
The bar graph compares average mental health days across different general health statuses. Individuals reporting 'Poor' general health experience significantly more mental health days compared to those with 'Excellent' health. This highlights the critical interplay between mental and general health, emphasizing the need for integrated approaches to address mental health challenges within populations facing health issues

Question:
Do heart attacks affect individuals with different general health statuses, and what factors might these differences can indicate about the risk of heart disease?

```python
import matplotlib.pyplot as plt

ht_attk_hp = heart1.groupby('GeneralHealth')['HadHeartAttack'].value_counts().unstack()

ht_attk_hp.plot.bar(color=['red', 'green'])
plt.title('Heart Attack Rate by General Health Status of an individual')
plt.xlabel('General Health Status')
plt.ylabel('Count of Heart Attacks')
plt.xticks(rotation = 56)
plt.legend(title = 'Had Heart Attack', labels=['No', 'Yes'])
plt.grid(axis='y', alpha = 0.56)
plt.show()
```

Observation:
The bar graph illustrates the relationship between general health status and the count of heart attacks. It indicates that individuals with poor or low general health may experience a frequent or higher count of heart attacks compared to those in better health or have a healty living. The individual that has a poorer health need preventative measures to support individuals with lower general health status, and gradually reduces the risk of heart disease.
